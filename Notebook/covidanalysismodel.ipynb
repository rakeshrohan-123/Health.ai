```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
datasets = pd.read_excel('covid.xls')
```

```python
datasets.head()
```

```python
datasets.columns
```

```python
datasets.drop(['id', 'age', 'sex', 'diabetes', 'hypertension', 'cvd', 'copd', 'cancer',
       'renal', 'ace', 'arbs', 'acearbs', 'ageclass', 'ageclass1', 'ageclass2', 'ageclass3', 'ageclass4',
       'ageclass5', 'ageclass6', 'ageclass7', 'severe'], axis=1)
```

```python
datasets['death']=datasets['death'].map({True:'Dead',False:'Alive'})
```

```python
datasets.head()
```

```python
datasets.covidseverity
```

```python
X=datasets.loc[:,['covidseverity','verysevere']].values
```

```python
Y = datasets.iloc[:, -1].values
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
X_Test
```

```python
Y_Test
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf = GradientBoostingClassifier(loss='exponential',n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0).fit(X_Train, Y_Train)
clf.score(X_Test, Y_Test)
```

```python
clf
```

```python
clf.score(X_Test, Y_Test)
```

```python
clf.predict([[2,1]])[0]
```

```python
y_pred=clf.predict(X_Test)
```

```python
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(Y_Test, y_pred) 
cm
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf,X_Test,Y_Test)
plt.title('Confusion Matrix')
```
